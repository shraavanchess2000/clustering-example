```python
import numpy as np
import random
from random import choices
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
```

**Generate Sample Dataset with Gaussian Mixture Distribution**

```python
population = [1, 2, 3]
weights = [0.5, 0.2, 0.3]

pi_samples = choices(population, weights, k=180)
```

```python
sigma = 1

m1_samples = np.random.multivariate_normal([0, 0], sigma*np.eye(2), size=pi_samples.count(1))
m2_samples = np.random.multivariate_normal([1, np.sqrt(3)], sigma*np.eye(2), size=pi_samples.count(2))
m3_samples = np.random.multivariate_normal([-1, np.sqrt(3)], sigma*np.eye(2), size=pi_samples.count(3))
```

```python
plt.figure(figsize=(9, 6))

plt.plot(m1_samples[:, 0], m1_samples[:, 1], 'bo')
plt.plot(m2_samples[:, 0], m2_samples[:, 1], 'yo')
plt.plot(m3_samples[:, 0], m3_samples[:, 1], 'go')

plt.title('Plot of y vs. x for each sample')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['m1_samples', 'm2_samples', 'm3_samples'])
```

```python
m_samples = np.concatenate((m1_samples, m2_samples, m3_samples), axis=0)
```

**K-Means Clustering (k = 3)**

```python
k_means_clustering = KMeans(n_clusters=3, random_state=0).fit(m_samples)
```

```python
k_means_clustering.cluster_centers_
```

**Hierarchical Clustering**

```python
hierarchical_clustering = AgglomerativeClustering(n_clusters=3).fit(m_samples)
```

```python
hierarchical_clustering_centers = np.empty((3, 2))

hierarchical_clustering_centers[0, 0] = np.mean(m_samples[np.where(hierarchical_clustering.labels_==0), 0])
hierarchical_clustering_centers[0, 1] = np.mean(m_samples[np.where(hierarchical_clustering.labels_==0), 1])
hierarchical_clustering_centers[1, 0] = np.mean(m_samples[np.where(hierarchical_clustering.labels_==1), 0])
hierarchical_clustering_centers[1, 1] = np.mean(m_samples[np.where(hierarchical_clustering.labels_==1), 1])
hierarchical_clustering_centers[2, 0] = np.mean(m_samples[np.where(hierarchical_clustering.labels_==2), 0])
hierarchical_clustering_centers[2, 1] = np.mean(m_samples[np.where(hierarchical_clustering.labels_==2), 1])

print(hierarchical_clustering_centers)
```

**EM Algorithm**

```python
em_clustering = GaussianMixture(n_components=3, random_state=0).fit(m_samples)
```

```python
print(em_clustering.means_)
```
